## Dictionary statistics


Dictionary statistics describe the structure of the dictionary.

The program displays statistics of the dictionary:

○ number of lines,

○ number of words,

○ number of unique words,

○ top 10 words with the most occurrences in the file.

○ letters (of a proper language) with counters of occurrences in the words and the other
characters found in words in the dictionary.

```python
stats = statistics.dictionary_stats("data/odm.txt")
print(f'Dictionary statistics:\n {stats}')
```

## Works statistics

Works statistics describe the linguistic properties of the analyzed texts.

The function displays statistics of the files:

○ number of files,

○ number of lines,

○ number of words,

○ number of unique words,

○ top 10 words with the most occurrences in the file

○ letters (of a proper language) with counters of occurrences in the words, then the other
characters found in files

If more than one file was given, the statistics are given as a total for all files, and also
for each separate file.

```python
work_stats = statistics.works_stats("data/pan_tadeusz.txt,data/oda-do-mlodosci.txt,data/niepewnosc.txt")
print(f'Files statistics:\n{work_stats}')
```

## Style similarity

The similarity between Master texts is calculated basen on the n most frequent words in each of them. The function similarity compares n most frequent words in two texts and calculates the part of all these words which overlap in these sets. The similarity score ranges from 0 to 100.

```python
dictionary = load.load_dictionary("data/odm.txt")
texts, _ = load.load_works("data/pan_tadeusz.txt,data/oda-do-mlodosci.txt,data/niepewnosc.txt")

df = analysis.similarity_comparison("data/pan_tadeusz.txt,data/oda-do-mlodosci.txt,data/niepewnosc.txt", texts, 10,"results")
```
